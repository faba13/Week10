```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("cust_seg.csv")
```

```python
# What kind of data does each column hold?

for col in df:
    print(col + ":", df[col].unique())
```

```python
# Looking at the code above, it is appropriate to remove all trailing spaces from the "age" and "antiguedad" columns. There is a method named strip to do this

df["age"] = df["age"].apply(lambda string: string.strip())
df["antiguedad"] = df["antiguedad"].apply(lambda string: string.strip())
```

```python
# Convert "NA" strings into the Pandas nan type

df["age"] = df["age"].replace("NA", np.nan)
df["antiguedad"] = df["antiguedad"].replace("NA", np.nan)
```

```python
# The "age" and "antiguedad" columns are encoded as strings, not numbers

def convert_to_int(data):
    try:
        return int(data)
    except:
        return data

df["age"] = df["age"].apply(convert_to_int)
df["antiguedad"] = df["antiguedad"].apply(convert_to_int)
```

```python
# What kind of data does each column hold?

for col in df:
    print(col + ":", df[col].unique())
```

Here we shall investigate the NA in the data.

```python
# How many NA are in each column?
for col in df:
    print(col + ":", df[col].isna().sum())
```

Notice that there is some strange data that is incorrect. For instance, the antiguedad (customer seniority) column includes a value of -999999. This is probably a placeholder for an unknown value, so we could replace this with nan.

```python
df["antiguedad"] = df["antiguedad"].replace(-999999, np.nan)
```

What to do with the NA values? Where the data is quantitative and continuous, we may assume the average or the median value to fill in the blanks. Where the data is categorical (even if numerically encoded), it will depend on the context. First, let's deal with the quantitative fields:

* age
* fecha_alta
* ult_fec_cli_1t
* antiguedad
* renta

Of these, only age is a continuous variable, making it appropriate to use the mean. For the rest, we shall use the median.

```python
for col in ["renta"]:
    df[col] = df[col].fillna(df[col].mean())
```

```python
for col in ["age", "antiguedad"]:
    df[col] = df[col].fillna(df[col].median())
```

```python
from datetime import datetime

def convert_date(string):
    try:
        return datetime.strptime(string, "%Y-%m-%d").timestamp()
    except:
        return string

df["fecha_alta"] = df["fecha_alta"].fillna(datetime.fromtimestamp(df["fecha_alta"].map(convert_date).median()).strftime("%Y-%m-%d"))
df["ult_fec_cli_1t"] = df["ult_fec_cli_1t"].fillna(datetime.fromtimestamp(df["ult_fec_cli_1t"].map(convert_date).median()).strftime("%Y-%m-%d"))
```

For the other fields:

* ind_empleado

We can use the mode since they are all categorical.

```python
for col in df:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
# How many NA are in each column?
for col in df:
    print(col + ":", df[col].isna().sum())
```

All NA have now been processed and replaced with an appropriate value. Now time to extract the transformed dataset.

```python
df.drop(["Unnamed: 0", "unrecorded_values"], axis=1).to_csv("cust_seg_transformed.csv")
```
